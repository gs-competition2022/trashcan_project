```python
import pandas as pd
import numpy as np
from numpy import log as ln


data = pd.read_csv('../DATA_P/행정동별.csv')
data = pd.DataFrame(data)

#접근성은 제외

data['정보화지수'] = data['전체인구']/data['전체인구'] + data['고령자 수']/data['전체인구']* 0.5 + data['장애인 수']/data['전체인구'] * 0.668 + data['기초 수급자 수']/data['전체인구']*0.853
data['실외 와이파이 비율'] = data['실외 와이파이 수']/data['와이파이 수']
data['전체인구'] = 1/data['전체인구']
data2 = data[['행정동','와이파이 수','실외 와이파이 비율','정보화지수','전체인구','평균소득']]

data = data[['와이파이 수','실외 와이파이 비율','정보화지수','전체인구','평균소득']]

dt = []

for i in range(20) :
    li = []
    for j in data.columns :
        li.append(list(data[j])[i])
    dt.append(li)

data1 = dt
```

```python
n = len(data1[0])
m = len(data1)

for i in range(n):
    
    cl = []
    
    for j in range(m):
        
        cl.append(data1[j][i])
    
    max_n = max(cl)
    min_n = min(cl)
    
    for j in range(m) :
        data1[j][i] = round((data1[j][i] - min_n) / (max_n - min_n),3)
        


entropy = []

k = 1 / ln(m)

for i in range(n) :
    
    e = []
    
    for j in range(m):
        
        d = data1[j][i]
        if ln(d) == "-inf"  :
            e.append(0)
        else:
            ld = round(ln(d),3)
            if np.isnan(d*ld):
                e.append(0)
            else:
                e.append(d*ld)
        
        
    entropy.append(k * sum(e))
    

weight = []

sum_ent = sum(entropy)

for i in range(n) :
    weight.append(entropy[i]/sum_ent)

print(weight)
```

```python
sod = pd.read_csv('../DATA/소득정보211203.csv')
sod.head()
l = list(sod['adstrd_nm'])
gu = []
hjd = []
sido = []
for i in l:
    i.split(' ')
    if i[0]:
        sido.append(i[:5])
    else :
        sido.append(' ')
    if i[1] :
        gu.append(i[6:9])
    else :
        gu.append(' ')
    if i[2]: 
        hjd.append(i[10:])
    else :
        hjd.append(' ')
sod['시'] = sido
sod['구'] = gu
sod['행정동'] = hjd
sod
```

```python
sod = sod[['행정동','ave_income_amt','tot_hshld_co']]
hjd_set = list(set(list(sod['행정동'])))
avg_income = []
for i in hjd_set :
    ic = 0
    h = 0
    for j in range(len(sod)):
        if i == list(sod['행정동'])[j]:
            ic += list(sod['ave_income_amt'])[j] * list(sod['tot_hshld_co'])[j]
            h+= list(sod['tot_hshld_co'])[j]
    avg_income.append(ic/h)
sode = pd.DataFrame()
sode['행정동'] = hjd_set
sode['평균소득'] = avg_income
sode
```
