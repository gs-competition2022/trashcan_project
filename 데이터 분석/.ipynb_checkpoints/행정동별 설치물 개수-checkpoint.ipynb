```python
import pandas as pd

d1 = pd.read_csv('../DATA_P/강서구 가로쓰레기통.csv')
d2 = pd.read_csv('../DATA_P/강서구 보안등.csv')
d3 = pd.read_csv('../DATA_P/강서구 와이파이.csv')
d4 = pd.read_csv('../DATA_P/강서구 정보소외.csv')
d5 = pd.read_csv('../DATA_P/강서구 스마트가로등.csv')
```

```python
d1['행정동'] = d1['Unnamed: 0']
d1['쓰레기통 수'] = d1['설치위치']
d1 = d1[['행정동','쓰레기통 수']]
d1.head()
```

```python
d2['행정동'] = d2['hjd']
d2['보안등 수'] = d2['num']
d2 = d2[['행정동','보안등 수']]
d2.head()
```

```python
d3['와이파이 수'] = d3['wifi_num']
d3['실내 와이파이 수'] = d3['wi']
d3['실외 와이파이 수'] = d3['wo']
d3 = d3[['행정동','와이파이 수','실내 와이파이 수','실외 와이파이 수']]
d3.head()
```

```python
d4['행정동'] = d4['동']
d4 = d4[['행정동','전체인구','고령자 수','장애인 수','기초 수급자 수']]
d4.head()
```

```python
d5['스마트가로등 수'] = d5['갯수']
d5 = d5[['행정동','스마트가로등 수']]
d5.head()
```
