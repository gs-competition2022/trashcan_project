# 정보취약층 분석

## 1. 강서구 행정동별 고령층 분석

### a) 데이터 import 후, 전처리

```python
import pandas as pd
#서울시 고령층 데이터 import 후, 강서구 데이터만 추출
df = pd.read_excel('../DATA/서울시 고령층 데이터 2021.xls',header = 1)
df = pd.DataFrame(df)
df = df[df['자치구']=='강서구']
df = df[['자치구','동','전체인구','합계']]
df = df.drop([df.index[0]])
#고령층 비율 값을 구함
df["고령층 비율"] = df["합계"] / df["전체인구"]
df["고령자 수"] = df["합계"]
df = df.drop(["합계"],axis=1)
```

### b) 행정동별 고령자 수 정렬

```python
df = df.sort_values(by=["고령자 수"], ascending=[False])
df
```

## 2. 강서구 행정동별 장애인 분석

### a) 데이터 import 후 전처리

```python
df1 = pd.read_excel('../DATA/서울시 장애인 데이터 2020.xls',header = 1)
df1 = pd.DataFrame(df1)
df1 = df1[df1['자치구']=='강서구']
df1['장애인 수'] = df1['합계']
df1 = df1[['자치구','동','장애인 수']]
df1 = df1.drop([df1.index[0]])
```

### b) 행정동별 장애인 수 정렬

```python
df1 = df1.sort_values(by=["장애인 수"], ascending=[False])
df1
```

## 3.강서구 행정동별 기초수급자 분석

### a) 데이터 import 후 전처리

```python
df2 = pd.read_excel('../DATA/서울시 기초수급자 정보 2021.xls',header = 1)
df2 = pd.DataFrame(df2)
df2 = df2[df2['자치구']=='강서구']
df2
df2 = df2[['자치구','동','인원']]
df2 = df2.drop([df2.index[0]])
df2 = df2.drop(df2[df2['동']=='기타'].index)
```

### b) 행정동별 기초 수급자 수 정렬

```python
df2 = df2.sort_values(by=["인원"], ascending=[False])
df2
```

## 4. 정보 취약 계층 총 분석

### a) 데이터프레임 취합

```python
df = df.sort_values(by=['동'])
df1 = df1.sort_values(by=['동'])
df2 = df2.sort_values(by=['동'])
df_x = pd.merge(df, df1, on ="동")
df_y = pd.merge(df_x, df2, on="동")
df_y['기초 수급자 수'] = df_y['인원']
fin_dt = df_y[['동','전체인구','고령자 수','장애인 수','기초 수급자 수']]
```

### b) 인원 총합 및 비율 구하기

```python
fin_dt['총합'] = fin_dt['고령자 수']+fin_dt['장애인 수']+fin_dt['기초 수급자 수']
fin_dt = fin_dt.sort_values(by=['총합'])
#중복이 있을 수 있으므로 총합은 고려하지 않는 것이 좋을듯
```
