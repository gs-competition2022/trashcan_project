```python
import pandas as pd

#가로 쓰레기통 데이터, 자치구단위 서울 생활인구 데이터 import.
#각각 dataframe화 하여 df, df1로 정의
df = pd.read_excel('../DATA/서울특별시 가로쓰레기통 현황_202106.xlsx',header = 1)
df1 = pd.read_csv('../DATA/자치구단위 서울생활인구 일별 집계표.csv', encoding='CP949')

# df에서 '자치구명' column 별로 몇개의 쓰레기통이 있는지 value_counts()로 집계한 dataframe g0 정의
g0 = pd.DataFrame(df['자치구명'].value_counts())

# df1에서'시군구명'column 별로 '총생활인구수' 평균을 구한 dataframe g 정의
g = pd.DataFrame(df1['총생활인구수'].groupby(df1['시군구명']).mean())

# g에서 index가 '서울시' 인 데이터는 필요없으므로 drop하여 제거 후, index 명으로 sort(전처리)
g = g.drop(['서울시']).sort_index()

# g와 g0의 index명을 동일하게 만들기 위해 g0의 '강서'를 g와 동일한 '강서구'로 변경 후, index 명으로 sort(전처리)
g0 = g0.rename(index={'강서':'강서구'}).sort_index()

# 전처리한 g와 g0의 값을 이용해 자치구별로 (쓰레기통 갯수)/(총생활인구수) 비율을 구한 dataframe tp_ratio를 구함. 
tp_ratio = pd.DataFrame(g0['자치구명']/g['총생활인구수'])

#column명을 '비율'로 설정
tp_ratio.columns = ['비율']

#값에 대해 정렬해서 결과값 출력
tp_ratio.sort_values(by=['비율'])
```
