```python
import numpy as np
from numpy import log as ln

fin = pd.read_csv('../DATA_P/행정동별 공공설치.csv')
fin = pd.DataFrame(fin)
fin['공공설치물 총합'] = 1/fin['공공설치물 총합']
fin1 = fin

fin = fin[['다중시설 총합','공공설치물 총합','전체인구']]
dt = []
for i in range(20) :
    li = []
    for j in list(fin.columns) :
        li.append(list(fin[j])[i])
    dt.append(li)

data1 = dt
n = len(data1[0])
m = len(data1)

for i in range(n):
    
    cl = []
    
    for j in range(m):
        
        cl.append(data1[j][i])
    
    max_n = max(cl)
    min_n = min(cl)
    
    for j in range(m) :
        data1[j][i] = round((data1[j][i] - min_n) / (max_n - min_n),3)
        


entropy = []

k = 1 / ln(m)

for i in range(n) :
    
    e = []
    
    for j in range(m):
        
        d = data1[j][i]
        if ln(d) == "-inf"  :
            e.append(0)
        else:
            ld = round(ln(d),3)
            if np.isnan(d*ld):
                e.append(0)
            else:
                e.append(d*ld)
        
        
    entropy.append(k * sum(e))
    

weight = []

sum_ent = sum(entropy)

for i in range(n) :
    weight.append(entropy[i]/sum_ent)

print(weight)
```
