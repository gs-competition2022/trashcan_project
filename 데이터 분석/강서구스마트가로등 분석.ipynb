```python
import pandas as pd
from collections import Counter
import folium
```

### 0. Data 준비

```python
# DataFrame 형태로 파일 읽어오기 
data = pd.read_csv('C:\\Users\\이찬희\\Desktop\\서울특별시_강서구_스마트가로등_20210319_1616992721862_24971.csv', encoding='CP949')
data
```

```python
# 열 종류 확인 
data.columns

#Index(['스마트가로등유형', '시도명', '시군구명', '도로명', '소재지도로명주소', '소재지지번주소', '위도', '경도',
#'스마트가로등형태', '점등상태여부', '센서종류', '센싱정보서비스', 'CCTV유무', 'WiFi유무', 'GPS수신기유무',
#'비콘유무', '조명제어여부', '위급상황신고가능여부', '앱서비스명', '앱서비스제공내용', '기타서비스', '설치연도',
#'관리기관명', '관리기관전화번호', '데이터기준일자'])
```

### 1. 교통 가로등 제외

```python
# 센싱이 도로 조명이 아닌 교통 조명인 경우 Drop
d = data[data['센싱정보서비스'].str.contains('교통')].index
data.drop(d, inplace = True)

data
```

##### 결과: row 개수 126 → 122 제거 완료 

### 2. 동 별로 COUNT 

```python
# 위도, 경도가 적혀있지 않은 경우는 map에 보이기에 적합하지 않아 drop
# 주소만 적혀있는 경우는 주소로만으로 추후에 사용할 수 있으니 새 변수 data_map을 생성해 사용
data_map = data.dropna(subset=['위도', '경도'])

# 강서구 지도 띄우기
m = folium.Map([37.55459659834, 126.83722951069 ],zoom_start=12)
```

```python
for i in data_map.index:
    sub_lat =  data_map.loc[i,'위도']
    sub_long = data_map.loc[i,'경도']
    
    #지도에 데이터 찍어서 보여주기
    tooltip = "스마트 가로등"
    folium.Marker([sub_lat, sub_long],  tooltip =tooltip).add_to(m) 
    
m
```

```python
dong = []

for i in data['소재지지번주소']:
    a = i.split(" ")
    dong.append(a[2])
    
print(dong)
```

```python
# 강서구 내 동 이름으로 스마트 가로등 Count
cnt_dong = Counter(dong)

print(cnt_dong)
```

##### 결과: {'방화동': 60, '공항동': 34, '마곡동': 22, '가양동': 6}
#####          → 방화동 쪽에 스마트 가로등이 모여있는 현상을 볼 수 있다

### 3. 스마트 가로등 Wifi 유무 파악

```python
cnt_wifi = Counter(data['WiFi유무'])

print(cnt_wifi)
```

##### 결과: {'N': 122} → 스마트 가로등에 Wifi가 설치된 스마트 가로등이 아직은 강서구에는 존재하지 않는다.

### 4. GPS 수신기 유무 파악

```python
cnt_GPS = Counter(data['GPS수신기유무'])

print(cnt_GPS)
```

#### 결과: {'N': 122} → 강서구에는 GPS수신기가 달린 스마트 가로등은 아직 존재하지 않는다.
